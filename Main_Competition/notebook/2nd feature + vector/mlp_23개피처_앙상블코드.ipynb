## Creating mean ensembles from submission files

```python
# 사용법: 
# 1) 스크립트를 실행하기 전에 submissions 폴더를 먼저 만듭니다. 
# 2) 앙상블할 submission 화일을 submissions 폴더에 저장합니다.
# 3) 실행하면 현재 폴더에 앙상블한 submission 화일이 생성됩니다.

import pandas as pd
import numpy as np
import os

# 아래 폴더가 있다는 가정 하에 수행
folder = 'mlp submissions'
nf = 0
for f in os.listdir(folder):
    ext = os.path.splitext(f)[-1]
    if ext == '.csv': 
        s = pd.read_csv(folder+"/"+f)
    else: 
        continue
    if len(s.columns) !=9:
        continue
    if nf == 0: 
        slist = s
    else: 
        slist = pd.merge(slist, s, on="ID")
    nf += 1

p = 0.001    # 이 값에 따라 성능이 달라짐 (p=1: 산술평균, p>1: 멱평균)
if nf >= 2:
    preds = []
    for i in range(8):
        pred = 0
        for j in range(nf):
            pred = pred + slist.iloc[:,j*8+i+1]**p 
        pred = pred / nf    
        pred = pred**(1/p)
        preds.append(pred)    
    preds = pd.DataFrame(preds).T
    preds.columns = ['F20','F30','F40','F50','M20','M30','M40','M50']
    pd.concat([pd.Series(slist.ID, name="ID"), preds] ,axis=1).to_csv('민수_cat+삐삐꺼.csv', index=False)
```
