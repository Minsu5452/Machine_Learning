```python
# Data Wrangling
import pandas as pd
import numpy as np
import os
```

```python
X_train = pd.read_csv('/content/drive/MyDrive/강민수+이효빈+한윤지/Input/X_train.csv', encoding = "cp949")
X_test = pd.read_csv('/content/drive/MyDrive/강민수+이효빈+한윤지/Input/X_test.csv', encoding = "cp949")
```

```python
X_train_id = X_train.custid
X_test_id = X_test.custid
```

```python
cat_features = X_train.select_dtypes(include=['object']).columns.to_list()
```

```python
X_train = pd.concat([X_train_id,X_train[cat_features]], axis =1)
X_test = pd.concat([X_test_id,X_test[cat_features]], axis = 1) 
```

```python
X_ALL = pd.concat([X_train, X_test],axis = 0)
```

```python
c = X_ALL.groupby('custid')['str_nm'].apply(lambda x: x.value_counts().index[0]).reset_index()
```

```python
df_c = c.iloc[:,0].reset_index(drop = True)
```

```python
df_c = pd.DataFrame(df_c)
```

```python
for i in cat_features:
  df_c = df_c.merge(X_ALL.groupby('custid')[f'{i}'].apply(lambda x: x.value_counts().index[0]).reset_index())
```
